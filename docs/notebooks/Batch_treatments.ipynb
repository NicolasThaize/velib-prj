```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns; sns.set()
```

```
# File location and type
file_location = "/FileStore/tables/velib_emplacement_des_stations.csv"
file_type = "csv"

# CSV options
infer_schema = "true" # inferSchema option is used to defined column type
first_row_is_header = "true"
delimiter = ";"

# The applied options are for CSV files. For other file types, these will be ignored.
df_stations = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df_stations)
```

```
df_stations = df_stations.toPandas()
```

```
df_stations = df_stations.rename(columns={"Identifiant station": "id_station", "Nom de la station": 'name', 'Coordonnées géographiques': 'coords'})
df_clean = df_stations.drop('Capacité de la station', axis=1)
df_clean.set_index('id_station', inplace=True)
df_clean.head()
```

```
df_clean.isna().sum() # Checking if na values
```

```
df_clean[['lat', 'long']] = df_clean['coords'].str.split(', ', expand=True) # Splitting coordinates to separates columns
df_clean['lat'] = pd.to_numeric(df_clean['lat']) # casting to numerical values
df_clean['long'] = pd.to_numeric(df_clean['long']) # casting to numerical values
df_clean.head()
```

```
X=df_clean.loc[:, ['lat', 'long']] # Data which is meant to be clustered
X.head()
```

```
# 30 clusters because of number of Paris district + nb of cities around
kmeans = KMeans(n_clusters = 74, init ='k-means++')
X['cluster_label'] = kmeans.fit_predict(X[X.columns[0:2]])
centers = kmeans.cluster_centers_ # Coordinates of cluster centers.
X.head(10)
```

```
X.plot.scatter(x = 'lat', y = 'long', c=X['cluster_label'], s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=50, alpha=0.5)
```

```
df_final = pd.DataFrame({'id_station':X.index, 'cluster_label':X['cluster_label']})
df_spark_final = spark.createDataFrame(df_final)
df_spark_final.head(10)
```

```
# Create a view or table

temp_table_name_2 = "station_clusters_csv"

df_spark_final.createOrReplaceTempView(temp_table_name_2)
```

```
%sql

/* Query to select the number of velib stations for each cluster */

select count(1) as count_custers
from station_clusters_csv
group by cluster_label;
```

```
%sql

/* Query to select the mean of velib places for each cluster */

select stt.cluster_label, round(mean(vlb.numdocksavailable), 0) as numdocksavailable_mean
from station_clusters_csv as stt
inner join (select stationcode, max(numdocksavailable) as numdocksavailable, is_installed
            from velib_raw_data_28_02_2023_csv
            group by stationcode, is_installed) as vlb on (vlb.stationcode = stt.id_station)
where (vlb.is_installed = 'OUI')
group by stt.cluster_label; 
```

```
%sql

/* Query to select the number of stations which sometimes have no velib available */

select count(1) as count_empty_stations
from station_clusters_csv
where (id_station in (select stationcode
                        from velib_raw_data_28_02_2023_csv 
                        where ((is_installed = 'OUI') and (numbikesavailable == 0))
                        group by stationcode));
```

```
%sql

/* Query to select the mean of velib places for each cluster */

select stt.id_station, vlb.name
from station_clusters_csv as stt
inner join (select stationcode, max(numdocksavailable) as numdocksavailable, name
            from velib_raw_data_28_02_2023_csv
            where ((is_installed = 'OUI') and (numbikesavailable == 0))
            group by stationcode, name) as vlb on (vlb.stationcode = stt.id_station); 
```
